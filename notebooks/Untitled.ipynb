### Random Forest Model

```python
# define X
X = df.drop(['YRMARR', 'BIRTHYR', 'YEAR', 'OCC', 'PERWT', 'STRATA', 'CLUSTER'], axis=1)
# define y
y = df.MARST
# define dummies
dummies = ['MIGRATE1', 'VETSTAT', 'EMPSTAT', 'ANCESTR1', 'STATEFIP', 'OCC_BROAD', 'CLASSWKR', 'NCHILD', 'MARRNO', 'NCHLT5', 
           'SEX', 'DEGFIELD', 'HISPAN', 'RACE', 'EDUC', 'MORTGAGE']
```

```python
# Import necessary modules
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

mapper = DataFrameMapper(
    [(d, LabelEncoder()) for d in dummies]
)


# Setup the pipeline
steps = [("mapper", mapper),("onehot", OneHotEncoder()), ('scaler', StandardScaler(with_mean=False)),
        ('rf', RandomForestClassifier(n_estimators=25, random_state=42, verbose=True, n_jobs=-1))]

# Create the pipeline: pipeline
pipeline = Pipeline(steps)

# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

# Fit the pipeline to the train set   
pipeline.fit(X_train, y_train)
```
